```python
#The histogram shows the distribution of laptop prices in  dataset.
#We can see which price ranges are most common and which are less common.
#Most Laptops are under 100k

sns.histplot(df['Price'], bins=10,kde=True)
plt.show()
```

```python
# The plot can helps to understand the relationship between the price and the rating of the laptops
# Laptops under 100k Rs have  rating less than 80

numeric_data.plot(kind='scatter', x='Price', y='Rating', s=32, alpha=.8)

plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Each scatter plot shows the relationship between price and rating for a specific brand
#Touch screen laptops have generally more price and ratings than other laptops
#Out of 9 , 4 brands do not have any touch screen laptops

top_brands = df['brand'].value_counts().index[:9]
df_top_brands = df[df['brand'].isin(top_brands)]
g = sns.FacetGrid(df_top_brands, col="brand", hue="is_touch_screen", height=4, aspect=1.2, col_wrap=3)
g.map(sns.scatterplot, "Price", "Rating", alpha=0.7)
g.add_legend()
g.set_axis_labels("Price", "Rating")
plt.show()
```

```python
# A series of histograms, each representing a different processor brand in  DataFrame.
# Each histogram shows the distribution of the number of threads for that processor brand.

g = sns.FacetGrid(df, col="processor_brand", col_wrap=3, height=4)
g.map(plt.hist, "num_threads")
g.fig.suptitle("Distribution of Number of Threads Across Different Processor Brands", y=1.02)
plt.show()
```

```python
# this is a scatterplot with regression line .
# this plot shows that the resolution width and resolution height are approximately proportional.

sns.lmplot(x="resolution_width", y="resolution_height", data=df)

plt.show()
```

```python
corr = numeric_data.corr()
corr
```

```python
f, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(corr,  cmap="coolwarm", vmax=.3, center=0,annot=True,
            square=True, linewidths=.5)

plt.title("Correlation Heatmap")
plt.show()
```

From the correlation matrix and Heatmap we can infer the following:

***Price and Specifications:*** The price of the laptop has a strong positive correlation with the number of cores, number of threads, RAM memory, and primary storage capacity. This suggests that laptops with higher specifications tend to be more expensive.

***Price and Display:*** The price has a moderate positive correlation with the display size and resolution width. This indicates that laptops with larger and higher resolution displays are generally more expensive.

***Processor and RAM:*** The number of cores and threads in the processor have a strong positive correlation with the RAM memory. This suggests that laptops with more powerful processors usually come with more RAM.
Resolution Width and Height: The resolution width and height are strongly positively correlated, which is expected as larger values for one typically imply larger values for the other.

***Secondary Storage:*** The secondary storage capacity does not seem to have a significant correlation with the price or other features. This could be because not all laptops have secondary storage, and the capacity may vary widely for those that do.

***Rating and Specifications:*** The laptopâ€™s rating has a strong positive correlation with the number of cores, number of threads, and RAM memory. This suggests that laptops with higher specifications tend to receive higher ratings.

***Resolution Height and Width*** The laptopâ€™s Resolution Height has a strong positive correlation with the Resolution Width.

```python
#The diagram shows the relationships between the three categories: â€˜processor_brandâ€™, â€˜processor_tierâ€™, and â€˜gpu_brandâ€™
#The width of the lines at each category level corresponds to the count of laptops with that combination of categories.
#This allows to see which combinations are most common in data.

fig2 = px.parallel_categories(df[['processor_brand', 'processor_tier', 'gpu_brand']],
                              title='Parallel Categories Diagram for Processor Brand, Processor Tier and GPU Brand')
fig2.show()
```
